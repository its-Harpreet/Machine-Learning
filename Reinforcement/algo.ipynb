```python
import pandas as pd
import thompson as tp
```

```python
url='https://raw.githubusercontent.com/digipodium/Datasets/main/reinforcement/Ads_CTR_Optimisation.csv'
df= pd.read_csv(url)
df.head()
```

```python
for col in df:
    count= df[col].value_counts()
    print(count)
    print('-'*50)
```

```python
out= tp.UCB(df)   #train reinforcement model
```

```python
tp.plot(out)
```

```python
re=tp.thompson(df)  #train with Thompson Sampling
```

```python
tp.plot(re)
```
